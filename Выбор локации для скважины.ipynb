## –û–±—É—á–µ–Ω–∏–µ –∏ –ø—Ä–æ–≤–µ—Ä–∫–∞ –º–æ–¥–µ–ª–∏

–í–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è –º–æ–¥–µ–ª—å—é –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ —Ç–∞–∫ –∫–∞–∫ –æ–Ω–∞ –ª—É—á—à–µ –≤—Å–µ–≥–æ –ø–æ–¥—Ö–æ–¥–∏—Ç –¥–ª—è –¥–∞–Ω–Ω–æ–π –∑–∞–¥–∞—á–∏ –∏ –æ—Ü–µ–Ω–∏–º –∫–∞—á–µ—Å—Ç–≤–æ –ø–æ–¥–µ–ª–∏ –ø–æ –º–µ—Ç—Ä–∏–∫–µ R2.

````python
def ML_linear(features_train,target_train,features_valid,target_valid,region):
    best_R2 = 0
    model = LinearRegression() 
    model.fit(features_train, target_train) 
    volume = model.predict(features_valid)
    R2 = model.score(features_valid,target_valid)
    best_R2 = R2 
    mse = mean_squared_error(target_valid, volume)
    print(f"{region} —Ä–µ–≥–∏–æ–Ω\nRMSE = {mse ** 0.5}")
    print(f"–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –æ–±—ä–µ–º–∞ —Å–∫–≤–∞–∂–∏–Ω—ã:{volume.mean()}+-{mse ** 0.5}")
    print(f"R2 –¥–ª—è {region} —Ä–µ–≥–∏–æ–Ω–∞ - {best_R2}\n")
    return model,volume,mse ** 0.5
````

````python
modelLR1,volume1,mse1 = ML_linear(features1_train,target1_train,features1_valid,target1_valid,1)
modelLR2,volume2,mse2 = ML_linear(features2_train,target2_train,features2_valid,target2_valid,2)
modelLR3,volume3,mse3 = ML_linear(features3_train,target3_train,features3_valid,target3_valid,3)
````

* –ß—Ç–æ–±—ã  –≤–∑—è—Ç—å –ø–µ—Ä–≤—ã–µ 200 –∑–Ω–∞—á–µ–Ω–∏–π —Ç–∞—Ä–≥–µ—Ç–∞ (–∫-–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—Ç –ª—É—á—à–∏–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è–º), –º–æ–∂–Ω–æ  `result` –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å –ø–æ —É–±—ã–≤–∞–Ω–∏—é –∏ –ø—Ä–∏–º–µ–Ω–∏—Ç—å  `[:200]`

–§—É–Ω–∫—Ü–∏—è profit –ø–µ—Ä–µ–¥–∞–µ—Ç –Ω–∞ –≤—ã—Ö–æ–¥ —Å—É–º–º–∞—Ä–Ω—É—é –ø—Ä–∏–±—ã–ª—å —Å –≤—ã–±—Ä–∞–Ω–Ω—ã—Ö 200 —Å–∞–º—ã—Ö –±–æ–ª—å—à–∏—Ö –ø–æ –æ–±—ä–µ–º—É —Å–∫–≤–∞–∂–∏–Ω.

–í–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è —Ç–µ—Ö–Ω–∏–∫–æ–π `Bootstrap —Å 1000 –≤—ã–±–æ—Ä–æ–∫`, —á—Ç–æ–±—ã –Ω–∞–π—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–∏–±—ã–ª–∏.

–î–∞–ª–µ–µ –Ω–∞–π–¥–µ–º `—Å—Ä–µ–¥–Ω—é—é –ø—Ä–∏–±—ã–ª—å`, `95%-–π –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª` –∏ `—Ä–∏—Å–∫ —É–±—ã—Ç–∫–æ–≤`, –∫–æ—Ç–æ—Ä—ã–π –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –º–µ–Ω—å—à–µ **2.5%**

````python
def characteristics(target,volume,BUDGET):
    state = np.random.RandomState(12345)
    lesion = []
    for i in range(1000):
        part_target = target.sample(n = 500, replace=True, random_state=state)
        volume = pd.Series(volume, name = 'predict')
        part_target = pd.Series(part_target, name = 'profits')
        volume_boot = volume.iloc[part_target.index] 
        part_target.reset_index(drop = True)
        volume_boot.reset_index(drop = True)
        boot = profit(part_target,volume_boot)
        lesion.append(boot-BUDGET)

    lesion = pd.Series(lesion)
    neg_count = len(list(filter(lambda x: (x < 0), lesion)))
    loss_probability = (neg_count/len(lesion))*100
    
    lower = lesion.quantile(0.025)
    upper = lesion.quantile(0.975)
    mean = lesion.mean()
    print("–†–∏—Å–∫–∏ —É–±—ã—Ç–∫–∞:", loss_probability)
    print(f"–°—Ä–µ–¥–Ω—è—è –ø—Ä–∏–±—ã–ª—å —Ä—É–±–ª–µ–π: {round(mean)} —Ä—É–±–ª–µ–π")
    print(f"95%-—ã–π –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª: {round(lower)} : {round(upper)} —Ä—É–±–ª–µ–π\n")
    return loss_probability, mean, lower, upper
````

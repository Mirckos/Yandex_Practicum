## Обучение

В этом параграфе будут рассмотрены три модели машинного обучения: 
- LinearRegression
- RandomForest
- CatBoost

Создадим функцию чтобы: 

- Дополнить признаков: дня и часа.

- Также сделаем ряд стационарным с помощью разности соседних значений ряда.

- Добавим в признаки прошлые значения ряда, для предсказания тендеции.

- Дополним признак скользящего среднего 

```python
def make_features(df, rolling_size , lag_m):
    df_c =df.copy()
    df_c["stat"] = abs(df_c['num_orders'] - df_c['num_orders'].shift())
    df_c['hour'] = df_c.index.hour
    df_c['dayofweek'] = df_c.index.dayofweek
    
    for lag in range(1, lag_m + 1):
        df_c[f'lag_{lag}'] = df_c['num_orders'].shift(lag)

    df_c['rolling_mean'] = df_c['num_orders'].shift().rolling(rolling_size).mean()
    return df_c
```

<div class="alert alert-block alert-info">

<b>Совет: </b> Желательно было написать функцию таким образом, чтобы она возвращала новый датафрейм, а не изменяла глобальную переменную.
1. Это хорошая практика (не изменять глобальные переменные внутри функции), которая улучшает читаемость кода.
2. В какой-то момент может появиться необходимость подбирать некоторые параметры используемый внутри функции. В этот момент копирование датасета станет обязательным.

Преобразуем данные и удалим пропуски:

```python
df = make_features(df, 50, 10)
df = df.dropna()
```

Функция для разбиения на две выборки - обучающую **(90%)** и тестовую **(10%)**.

```python
def prepare_ML(df,targets):
        target = df[targets]
        features  = df.drop(targets,axis = 1)
        
        features_train,features_check,target_train,target_check  = train_test_split(features, target ,shuffle=False, test_size=0.1)

        return features_train,features_check,target_train,target_check
```

```python
features_train,features_test,target_train,target_test = prepare_ML(df,"num_orders")
```

<div class="alert alert-block alert-success">
<b>Успех:</b> Разбиение было сделано корректно.
</div>

Проверим соотношение выборок:

```python
full = (features_train.shape[0]+features_test.shape[0])
print(f"Процент обучающей выборки/тестовой выборки: \
{(features_train.shape[0]*100/full):.1f}/{(features_test.shape[0]*100/full):.1f}")
features_train
```

Отмашстабируем данные:

```python
scaler = StandardScaler()
scaler.fit(features_train[features_train.columns])

features_train[features_train.columns] = scaler.transform(features_train[features_train.columns])
features_test[features_test.columns] = scaler.transform(features_test[features_test.columns])
features_test 
```

Создадим score для применения в моделях:

```python
def RMSE(target, predict):
    return (mean_squared_error(target, predict))**0.5

RMSE = make_scorer(RMSE, greater_is_better=False)
```

<div class="alert alert-block alert-success">
<b>Успех:</b> Отлично, что scaler был обучен только на треине! Скорер определен корректно.
</div>

### LinearRegression

Для применения cross_val_score на временных рядах нужно использовать TimeSeriesSplit.

```python
%%time
CV = TimeSeriesSplit(n_splits=5).split(features_train)

LR = LinearRegression()

LR.fit(features_train, target_train)

score = -pd.Series(cross_val_score(LR,features_train,target_train,scoring=RMSE,cv=CV)).mean()
print('RMSE на кросс-валидации:',score )
```

### RandomForest

```python
%%time
CV  = TimeSeriesSplit(n_splits=5).split(features_train)

RF = RandomForestRegressor()

parametrs = { 'n_estimators': range(5, 30, 5),
                     'max_depth': range(1, 15, 2), }
 
RF_GSCV = GridSearchCV(RF, parametrs, scoring=RMSE, cv=CV)
RF_GSCV.fit(features_train, target_train)
```

```python
%%time
RF_best_gp = RF_GSCV.best_params_
RF_rmse= -RF_GSCV.best_score_
print("Лучшие гиперпараметры:",RF_best_gp)
print("RMSE best:",RF_rmse)
```

### CatBoost

```python
%%time

CV = TimeSeriesSplit(n_splits=5).split(features_train)

CB = cb.CatBoostRegressor(task_type="GPU")

parametrs = {'learning_rate': [0.05,0.1],
                'depth': [4, 6, 8, 10]}

CB_GSCV = GridSearchCV(CB, parametrs, scoring=RMSE, cv=CV)
CB_GSCV.fit(features_train, target_train)
```

```python
%%time

CB_best_gp = CB_GSCV.best_params_
CB_rmse = -CB_GSCV.best_score_
print(CB_best_gp)
print(CB_rmse)
```

Найдены лучшие гиперпараметры для каждой модели и рассмотрены RMSE:
- CatBoost: 21.0
- LinearRegression: 31.4
- RandomForest: 21.8

Можно сказать, что лучшее всего себя показала модель **CatBoost**, проверим ее на тестовой выборке в следующем параграфе.

## Тестирование

Проведем тестирование модели на лучших гиперпараметрах.   

```python
%%time
CB = cb.CatBoostRegressor(depth = 6, learning_rate = 0.05)
CB.fit(features_train, target_train)
CB_predictions = CB.predict(features_test)

RMSE_cb = (mean_squared_error(target_test, CB_predictions))**0.5 
print('RMSE модели CatBoostRegressor:', RMSE_cb)
print('Общее время предсказания и обчения:')
```

Модель CatBoost  с заданными параметрами имеет RMSE = 38.49, что `подохит по требованиям заказчика`!

Рассмотрим погрешность между предсказанием и реальным значением на графике.

Рассмотрим признаки по важности и изобразим на графике:

```python
features_imp = list(CB.feature_importances_)
columns = features_train.columns
positive_imp = []
name_features = []

for x in features_imp:
    if x>0:
        positive_imp.append(x)
        name_features.append(columns[features_imp.index(x)])
    
plt.figure(figsize=(10, 10))
plt.title("Важность признаков")
 
plt.bar(range(len(positive_imp)), positive_imp)
plt.xticks(range(len(positive_imp)), name_features)
 
plt.tight_layout()
plt.show()
```

Самыми важными признаками для модели явлются "stat","hour","lag_1".
